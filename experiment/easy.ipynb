```python
import easyocr
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('CropNumbers/NumBase/20.bmp')
```

```python
plt.imshow(img)
```

```python
reader = easyocr.Reader(['en','ru'])
```

```python
def easy_find_test(img):
    result = reader.readtext(img)
    ans = []
    for i in result:
        ans.append(i[-2])
    return ans
```

```python
easy_find_test(img)
```
